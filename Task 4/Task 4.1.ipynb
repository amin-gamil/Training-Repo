```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
weather = pd.read_csv('Weather Dataset.csv')
weather.drop_duplicates(inplace=True,ignore_index=True)
weather.dropna(inplace=True,ignore_index=True)
for i  in weather.index:
   Temp = weather.loc[i, 'Temperature (C)']
   if (Temp > 100) or (Temp < -100):
     weather.drop(i,inplace=True)
for i  in weather.index:
    P = weather.loc[i, 'Pressure (millibars)']
    if (P < 870):
       weather.drop(i,inplace=True)
for i  in weather.index:
    H = weather.loc[i, 'Humidity']
    if (H < 0) or (H>1):
       weather.drop(i,inplace=True)
weather.drop('Loud Cover',axis=1,inplace=True)
weather['Formatted Date'] = pd.to_datetime(weather['Formatted Date'])
weather.sort_values('Formatted Date',inplace=True,ignore_index=True)
```

```python
xvalues = weather['Formatted Date']
yvalues = weather['Temperature (C)']
plt.plot(xvalues,yvalues)
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid(True)
```

```python
plt.hist(weather['Temperature (C)'],'auto')
```

```python
plt.scatter(weather['Temperature (C)'],weather['Humidity'],s=20)
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity')
```

```python
weather.corr(numeric_only=True)
import seaborn as sb
sb.heatmap((weather.corr(numeric_only=True)),cmap="coolwarm", annot=True)
```

A corelation coefficient for two variables indicates how strongly they affect each other.

*   A coefficient of +1 indicates a strong direct relation (i.e: both variables increase or decrease together).
*   A coefficient of -1 indicates a strong inverse relation.
*  A value of 0 indicates no relation between the 2 variables.

The heatmap shows the corelation matrix for evey variable in the dataset, with more saturated colors indicating stronger relationships. (Warm red for $+ve$ and Cool blue for $-ve$).
